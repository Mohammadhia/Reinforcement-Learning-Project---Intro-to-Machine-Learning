# I. Overview

The objective of this assignment is to implement reinforcement learning on an agent of Grid World. The goal of the agent is to be able to find the optimal path from any start point to the goal point through many iterations of trial and error, in which the optimal path is the one that results in the highest net reward. Grid World involves 3 types of states/spaces: Open spaces, which the agent can traverse freely; Wall/Obstacle spaces, which the agent should NOT go through; and the Goal space, which is where the agent is trying to reach. The agent is allowed to move left, right, up, and down from any space, with each movement resulting in the loss of 1 point from the net reward score; any landing on a wall/obstacle, out of bounds location, or invalid input resulting in a loss of 5 points from the net reward score; and the landing on the goal space resulting in a gain of 30 points to the net reward score.

The reinforcement learning solution was implemented using a type of Temporal Difference learning (either Q-learning or SARSA learning). The algorithm used also involved an epsilon greedy component, which led the agent to choosing the state with the highest immediate reward (or lowest cost) when determining which successive state to select.

# II. Maze Problem

The maze problem involves training an intelligent agent in finding an optimal path from a given start point to a predetermined goal state. The maze through which the agent must traverse to reach the desired goal contains obstacles like walls, which the agent must navigate around in its final optimal path. The optimal path is determined through hundreds of iterations, in which the net rewards for each action for each cell in the maze is calculated accordingly. Once the net rewards are calculated and stored in a data structure indicating the optimal action to take for each given state, the agent only needs to reference that in order to determine its optimal path to the goal state from a given input cell.

## Grid World (Sample Grid Environment File)

Grid World is an environment that represents the maze problem. In it, an agent attempts find the optimal path from an inputed start point to the predetermined goal point. The agent can move left, right, up, or down from a given cell, with a loss of 1 point for a movement to another open cell (to discourage unnecessary movements); a loss of 5 points for a movement into an obstacle cell, out of bounds, or for an invalid input; and a gain of 30 points for reaching the goal state.

The GridWorld class contains a series of methods to modify the position of the agent and obtain information regarding its movements and rewards, among other things. the __init__() function instantiates a GridWorld environment from a given text file containing a specific character to describe each cell (which can all be seen below). The get_cur_state() function returns the ordered pair describing where the agent is on the grid, and the goal_reached() function (which I added) returns whether or not the goal has been reached. The get_size() function returns the dimensions of the grid, the read_map() function is used to initialize the grid in __init__(), and the print_map() function prints out the grid. The check_state() method is used to determine if a state is valid, the init() function allows for the user to specify the initial state for the agent, and the next() function determines the reward for the agent given an action. Finally, the get_actions() function returns the list of possible actions that the agent can make.


Simple text file with three characters, 'O', 'H', and 'G'.
- 'O': open space
- 'H': Wall or obstacles
- 'G': Goal location

```python
# maze example

debug = False
verbose = False


class GridWorld:
    """ Grid World environment
            there are four actions (left, right, up, and down) to move an agent
            In a grid, if it reaches a goal, it get 30 points of reward.
            If it falls in a hole or moves out of the grid world, it gets -5.
            Each step costs -1 point. 

        to test GridWorld, run the following sample codes:

            env = GridWorld('grid.txt')

            env.print_map()
            print [2,3], env.check_state([2,3])
            print [0,0], env.check_state([0,0])
            print [3,4], env.check_state([3,4])
            print [10,3], env.check_state([10,3])

            env.init([0,0])
            print env.next(1)  # right
            print env.next(3)  # down
            print env.next(0)  # left
            print env.next(2)  # up
            print env.next(2)  # up

        Parameters
        ==========
        _map        ndarray
                    string array read from a file input
        _size       1d array
                    the size of _map in ndarray
        goal_pos    tuple
                    the index for the goal location
        _actions    list
                    list of actions for 4 actions
        _s          1d array
                    current state
    """
    def __init__(self, fn):
        # read a map from a file
        self._map = self.read_map(fn)
        self._size = np.asarray(self._map.shape) #Holds dimensions of the map [rows columns]
        self.goal_pos = np.where(self._map == 'G') #Holds coordinates of goal position [Like this: (array([5], dtype=int64), array([6], dtype=int64))]

        # definition of actions (left, right, up, and down repectively)
        self._actions = [[0, -1], [0, 1], [-1, 0], [1, 0]]
        self._s = None ### NOTE: self._s WILL ONLY EVER BE 'O' OR 'G' 

    def get_cur_state(self):
        return self._s
    
    def goal_reached(self): #Returns whether or not the goal has been reached (I ADDED THIS TO KNOW IF TO STOP ITERATIONS EARLY)
        return self.check_state(self._s) == 'G'

    def get_size(self):
        return self._size

    def read_map(self, fn): #Called in __init__ method to read text file containing the Grid World
        grid = []
        with open(fn) as f:
            for line in f:
               grid.append(list(line.strip()))
        return np.asarray(grid)
    
    def print_map(self): #Prints the map as an array
        print( self._map )

    def check_state(self, s):
        if isinstance(s, collections.Iterable) and len(s) == 2:
            #If either index of the state is out of bounds for the grid, return 'N'
            if s[0] < 0 or s[1] < 0 or\
               s[0] >= self._size[0] or s[1] >= self._size[1]:
               return 'N'
            return self._map[tuple(s)].upper() #Otherwise, return the letter associate with the location of the state
        else:
            return 'F'  # wrong input

    def init(self, state=None): #Initializes according to a specified starting point (if no starting point is given, [0, 0] is default)
        if state is None:
            s = [0, 0]
        else:
            s = state

        if self.check_state(s) == 'O':
            self._s = np.asarray(state) #If s is an open space ('O'), then set self._s to that state
        else:
            raise ValueError("Invalid state for init")

    def next(self, a):
        s1 = self._s + self._actions[a] #Moves the state according to the action
        # state transition
        curr = self.check_state(s1) 
        
        if curr == 'H' or curr == 'N': #If move to a wall/obstacle or out of bounds, return -5 to total score
            return -5
        elif curr == 'F': #If input is invalid, return -5 to total score
            warnings.warn("invalid state " + str(s1))
            return -5
        elif curr == 'G': #If move results in move to goal, return 30 to total score and set current state to the new state
            self._s = s1
            return 30
        else: #Return -1 to total score if move to open state ('O')
            self._s = s1
            return -1
            
    def get_actions(self):
        return self._actions
```

Temporal Difference learning involves an implementation similar to the Value Iteration algorithm, in which an optimal Markov decision process (MDP) is found by having the agent act optimally for the transition to the next state as well as for all transitions in the future. The main difference in TD learning is that the agent doesn't have full knowledge of the entire MDP, so it needs to explore it over many iterations and update estimates of state values according to what it observes regarding transitions and rewards. For this assignment, TD learning could be implemented in 1 of 2 ways: State-Action-Reward-State-Action (SARSA) or Q-learning. In both ways, each state has a set of Q-Values for each possible action, in which the action associated with the maximum Q-value should be selected for each cell in the final policy.
- SARSA uses its current policy and Q-value of the next state to update the Q-value of a given state. This makes it "on-policy" learning.
- Q-learning uses an epsilon-greedy policy and Q-value of the next state to update the Q-value of a given state. This makes it "off-poicy" learning.

For this assignment, I chose to use Q-learning because it had the ability to find the optimal path independent of the policy used in traversing and updating the Q-values of the states. In it, I set the total number of epochs to 100 and the total number of iterations per epoch to 400. I included an epsilon-greedy function to allow the agent to select the action with the highest Q-value for its next successive action. Throughout this process, on each epoch, the first state is initialized. Then, for each iteration (step through the grid), the agent selects its next action based on the epsilon-greedy function. It uses the formula Q(s,a) = Q(s,a) + alpha[reward + gamma\*max(Q(s',a')) - Q(s,a)] to update the Q-value of each possible action for the state. It then updates the current state to the next state and repeats the process.

The epsilon_greedy function used in this assignment takes in the Q-table as well as the current state and epsilon value to determine which action to take. The epsilon value is used to ensure some randomness in selecting the next action (to allow for exploration), but it's relatively low so that the action with the highest Q-value for the given state is usually selected. This is a greedy algorithm because it only selects the immediate highest value action in the hopes of finding the global optimal path by continuing this process.

Below, I set the startState to the point (0,0), the learning rate (alpha) to 0.05 so as to not make it too big and miss the optimum value, the discount factor (gamma) to 0.9 to place a fairly high emphasis on successive states, and the epsilon value to 0.1 to allow for the slight chance that a random action will be selected, allowing for exploration (otherwise, selecting the action with the highest Q-value). The number of iterations per epoch is set to 400, the number of states is set to the number of rows * number of columns in GridWorld, and the number of actions is attained from the get_actions method. The e_greedy function uses the epsilon greedy method discussed above and defaults to the greedy method when not choosing a random action.

```python
startState = [0,0] #Will hold the starting point
alpha = 0.05  #Learning rate
gamma = 0.9 #Discount factor
epsilon = 0.1

n_iterations = 400 #Number of iterations
n_states = env.get_size()[0] * env.get_size()[1] #Number of possible states is equal to number of rows * number of columns
n_actions = len(env.get_actions())

# action selection
def greedy(Q, s):
    maxValues = np.argwhere(Q[s,:] == max(Q[s,:])) #Makes an array of all indices with the highest value (in case there's a tie)
    maxValues = maxValues.flatten().tolist()
    return rnd.choice(maxValues) #Randomly chooses action from list of max values
        
def e_greedy(Q, s, e):   
    if np.random.rand() < e:  #Returns random action from list of possible actions if random value is less than epsilon
        return rnd.choice(possible_actions[s])
    else:
        return greedy(Q,s)
```

Below, a 2D Q-table is built with each cell being a state from GridWorld (the rows of the Q-table 2D array) and each column being the Q-value associated with the corresponding action. The initial values are set to 0.0 for each action per cell.

```python
#Generates list of lists (each inner list contains possible actions at current cell)
possible_actions = [] #Will hold the possible actions that can be taken per cell
for i in range(env.get_size()[0]):
    for j in range(env.get_size()[1]):
        possible_actions.append([0, 1, 2, 3])
        
Q = np.full((n_states, n_actions), -np.inf) #-inf for impossible actions (will stay as -inf after updating all cells if impossible)
for state, actions in enumerate(possible_actions):
    Q[state, actions] = 0.0 #Initial value of 0.0 for all possible actions for a given state/cell
```

The following code represents the actual Q-learning algorithm. Below are also a list of the net rewards that has a net reward appended per epoch and a number of steps list that records the number of iterations it took to reach the goal state (if it is reached) per epoch. The outer for loop goes through the number of epoch and initializes the start state according to what was listed above. The iterations are carried out accordingly.

The index of the current state is calculated to the corresponding row in the Q-table and from there, the next action to take (left, right, up, down) is found using the e_greedy() function. The next state is found by taking the next action and using it on the current state, and the reward is found by getting whatis returned from the next() function. The index of the new state is calculated the same way as it was for the first state, and it is used to update the Q-value of the first state and corresponding action, using the formula discussed previously. If the new state is out of bounds, the current state is the new state. If the goal is reached before the iterations are complete, the loop breaks.

The Q-table will be printed in the end (in the Results Section), and the final policy will be outputted as a 2D grid (like GridWorld) with each state receiving the number for the action with the highest Q-value in the Q-table.

```python
sumOfRewardsList = [] #Holds list of rewards for each epoch
numberOfStepsList = [] #Holds list of number steps to reach goal (which may not be reached) for each epoch

for e in range(100): #Number of epochs
    env.init(startState)

    sumOfRewards = 0.0 #Initial value of total reward sum for each epoch
    totalIterations = 0 #Initial value of total iteratons for each epoch
    
    for iteration in range(n_iterations):
        #Index of current cell (if 2D array was reshaped)
        currentStateIndex = (env.get_cur_state()[0] * env.get_size()[1]) + env.get_cur_state()[1] #(current row * number of columns) + current column
        a = e_greedy(Q, currentStateIndex, epsilon) #Action choice from epsilon greedy algorithm

        #Index of future possible cell (if 2D array was reshaped)
        sp = env.get_cur_state() + env.get_actions()[a] #Gets the new possible state using the action a (may not work due to hitting obstacle)
        reward = env.next(a) #Reward for action
        
        ##################### VALUES FOR GRAPHS IN RESULTS #####################
        sumOfRewards += reward #Updates the total sum
        totalIterations = iteration #Will update with the loop
        ########################################################################
        
        spIndex = (sp[0] * env.get_size()[1]) + sp[1] #(row of new possible state * number of columns) + column of new possible state
        if(spIndex > 62 or spIndex < 0): #If new state is out of bounds because of action, set new state equal to current state
            spIndex = currentStateIndex
                
        Q[currentStateIndex, a] += alpha * (reward + gamma * np.max(Q[spIndex, :]) - Q[currentStateIndex, a]) #Update Q-Value

        if(env.goal_reached() == True): #If goal state is reached
            break
            
    sumOfRewardsList.append(sumOfRewards)
    numberOfStepsList.append(totalIterations)
```

The values for $\gamma$, $\alpha$, and $\epsilon$ were chosen by using another for loop to the Q-learning algorithm listed above. 3 possible values were used for each parameter: [0.05, 0.1, 0.2] for alpha, [0.9, 0.85, 0.8] for gamma, and [0.1, 0.15, 0.2] for epsilon.
```python
alpha = [0.05, 0.1, 0.2]  #Learning rate
gamma = [0.9, 0.85, 0.8] #Discount factor
epsilon = [0.1, 0.15, 0.2]


totalSumOfRewardsList = []
totalNumberOfStepsList = []
for trial in range(9): #Trials to find optimal alpha, gamma, and epsilon
    
    ####################Generates list of lists (each inner list contains possible actions at current cell)####################
    possible_actions = [] #Will hold the possible actions that can be taken per cell
    for i in range(env.get_size()[0]):
        for j in range(env.get_size()[1]):
            possible_actions.append([0, 1, 2, 3])

    Q = np.full((n_states, n_actions), -np.inf) #-inf for impossible actions (will stay as -inf after updating all cells if impossible)
    for state, actions in enumerate(possible_actions):
        Q[state, actions] = 0.0 #Initial value of 0.0 for all possible actions for a given state/cell
    ###########################################################################################################################
    if(trial >= 0 and trial <= 2):
        alphaIndex = trial % 3
        gammaIndex = 0
        epsilonIndex = 0
    elif(trial >= 3 and trial <= 5):
        gammaIndex = trial % 3
        alphaIndex = 0
        epsilonIndex = 0
    elif(trial >= 6 and trial <= 8):
        epsilonIndex = trial % 3
        alphaIndex = 0
        gammaIndex = 0
        
    sumOfRewardsList = [] #Holds list of rewards for each epoch
    numberOfStepsList = [] #Holds list of number steps to reach goal (which may not be reached) for each epoch
    
    for e in range(100): #Number of epochs
        env.init(startState)

        sumOfRewards = 0.0 #Initial value of total reward sum for each epoch
        totalIterations = 0 #Initial value of total iteratons for each epoch

        for iteration in range(n_iterations):
            #Index of current cell (if 2D array was reshaped)
            currentStateIndex = (env.get_cur_state()[0] * env.get_size()[1]) + env.get_cur_state()[1] #(current row * number of columns) + current column
            a = e_greedy(Q, currentStateIndex, epsilon[epsilonIndex]) #Action choice from epsilon greedy algorithm

            #Index of future possible cell (if 2D array was reshaped)
            sp = env.get_cur_state() + env.get_actions()[a] #Gets the new possible state using the action a (may not work due to hitting obstacle)
            reward = env.next(a) #Reward for action

            ##################### VALUES FOR GRAPHS IN RESULTS #####################
            sumOfRewards += reward #Updates the total sum
            totalIterations = iteration #Will update with the loop
            ########################################################################

            spIndex = (sp[0] * env.get_size()[1]) + sp[1] #(row of new possible state * number of columns) + column of new possible state
            if(spIndex > 62 or spIndex < 0): #If new state is out of bounds because of action, set new state equal to current state
                spIndex = currentStateIndex

            Q[currentStateIndex, a] += alpha[alphaIndex] * (reward + gamma[gammaIndex] * np.max(Q[spIndex, :]) - Q[currentStateIndex, a]) #Update Q-Value

            if(env.goal_reached() == True): #If goal state is reached
                break

        sumOfRewardsList.append(sumOfRewards)
        numberOfStepsList.append(totalIterations)
    totalSumOfRewardsList.append(sumOfRewardsList)
    totalNumberOfStepsList.append(numberOfStepsList)

    policy = np.argmax(Q, axis=1)
    policy = np.reshape(policy, (env.get_size()[0], env.get_size()[1])) #Puts policy of what to do in each cell in easy to view grid
```
